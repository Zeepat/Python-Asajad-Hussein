```python
import matplotlib.pyplot as plt
import numpy as np
def math_func(gradient, power, y_intersect):
    function_list = []
    range_list = list(range(-10, 11))
    label_1 = str(gradient) + "x^" + str(power) + str(y_intersect)
    for i in range_list:
        function_list.append(int(gradient) * i ** int(power) + int(y_intersect))
    plt.plot(range_list, function_list, label=label_1)
    plt.legend()

math_func(1, 2, -3)
math_func(4, 1, -7)
        
```

```python
def change_calculator(sum_money, current_value):
    if (sum_money // current_value) != 0:
        antal = sum_money // current_value
        print(f'{antal}st {current_value}-lapp')

def change(sum_money):
    range = [1000, 500, 200, 100, 50, 20, 10, 5, 2, 1]
    for i in range:
        change_calculator(sum_money, i)
        sum_money = sum_money % i
        
money = int(input("Enter your total sum: "))
change(money)
```
