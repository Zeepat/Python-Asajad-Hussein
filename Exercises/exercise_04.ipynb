# Exercise 5

**a)**

```python
import random
dice_rolls_100 = range(100)
dice_roll_100_list = [random.randint(1, 6) for i in dice_rolls_100]
print(dice_roll_100_list)
counter = 0
for i in dice_rolls_100:
    if dice_roll_100_list[i] == 6:
        counter += 1

print(f"The number of outcome six in 100 dice rolls is: {counter}")
```

**b)**

```python
import random

dice_rolls = range(10**6)
dice_roll_list = [number * 0 + random.randint(1,6) for number in dice_rolls]

counter = 0
counter_list = list(range(6))

randoms = []
for i in range(6):
    for j in dice_roll_list[10**i:10**(i+1)]:
        if j == 6:
            counter += 1
            counter_list[i] = counter
        else:
            randoms.append(j)

print(f"Count number of six: {counter_list}")

accuracy_list = [counter_list[i]/10**(i+1) for i in range(6)]
accuracy_list_rounded = [round(i, 4) for i in accuracy_list]

print(f"P(six) = {accuracy_list_rounded}")
```

**a)**

```python
import random
import math
import matplotlib.pyplot as plt
points_range = range(5000)
points_list_x = [random.uniform(-1, 1) for number in points_range]
points_list_y = [random.uniform(-1, 1) for number in points_range]
distance_list = []
for i in points_range:
    distance_list.append(math.sqrt(points_list_x[i] ** 2 + points_list_y[i] ** 2))

for i in points_range:
    if distance_list[i] > 1:
        plt.scatter(points_list_x[i], points_list_y[i], color="orange")
    else:
        plt.scatter(points_list_x[i], points_list_y[i], color="blue")
```

**b)**

```python
import random
import math
import matplotlib.pyplot as plt

points_range = range(10**7)
points_list_x = [random.uniform(-1, 1) for number in points_range]
points_list_y = [random.uniform(-1, 1) for number in points_range]
distance_list = []
counter_outer = 0
counter_inner = 0

for i in points_range:
    distance_list.append(math.sqrt(points_list_x[i] ** 2 + points_list_y[i] ** 2))

for i in points_range:
    if distance_list[i] > 1:
        counter_outer += 1
    else:
        counter_inner += 1

fraction_inner = counter_inner/len(points_range)
print(fraction_inner)
```

The number converges closer and closer to pi/4. This has to do with how the area of a circle and the area of a square is calculated. You take the radius, which is 1, and square it, times pi gives us pi. The are of the square is calculated by taking one side and squaring it, such as -1 to 1 is 2, thus 2 squared is 4. Then taking the area of the circle divided by the total area, which is the squares area, gives us pi/4. This would be true for any circle that has the same diameter as the squares side.
